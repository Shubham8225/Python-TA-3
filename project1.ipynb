```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
m=np.mean(df,axis=1)
print("average of players in  matches\n\n",m)
```

```python
print("standard deviation ")
d=round((df.std(axis = 1, numeric_only = True)),10)
# print(d.max())
print(d)
```

```python

print("coeficient of variation = \n")
for i in range (len(d)):
  print(round((((d[i])/m[i])),4))

print("-----------")
print(d[1])
print(m[1])
```

```python
upper_limit = df['match_5'].mean() + 3*df['match_5'].std()
lower_limit = df['match_5'].mean() - 3*df['match_5'].std()
df['match_5'] = np.where(
    df['match_5']>upper_limit,
    upper_limit,
    np.where(
        df['match_5']<lower_limit,
        lower_limit,
        df['match_5']
    )
)
df['match_5'].describe()
```
